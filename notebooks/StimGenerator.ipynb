# Make random stimulus and blank order 

```python
import numpy as np 
import matplotlib.pyplot as plt

def plot_cycle(ordered_pairs, title):
    time = np.cumsum(ordered_pairs[:,1]/3) + np.arange(len(ordered_pairs[:,1])) + 1
    signal = ordered_pairs[:,0]

    fig, ax = plt.subplots(4,figsize = (17,10))
    ax[0].stem(time, signal)
    ax[1].stem(time, signal==2)
    ax[2].stem(time, signal==3)
    ax[3].stem(time, signal==5)

    ax[0].set_title(title)
    ax[1].set_title('2 dots only')
    ax[2].set_title('3 dots only')
    ax[3].set_title('5 dots only')
    ax[3].set_xlabel('time, volumes')
    ax[0].set_ylabel('Stimulus')
    ax[0].set_xticks([])
    ax[1].set_xticks([])
    ax[2].set_xticks([])
    ax[1].set_yticks([])
    ax[2].set_yticks([])
    ax[3].set_yticks([])
```

```python
# types of stimuli you want to include, 
# in the exact number of how many of them you want 
stim = [2,2,2,3,3,3,5,5,5]
# what durations of blanks do you want? 
# durations only, one of each type ( random number of each will be picked )
blanks = [15, 18, 21, 27]



pairs = []
for s in stim:
    pair = [s, np.random.choice(blanks)]
    pairs.append(pair)

order = np.arange(9)
np.random.shuffle(order)

pairs = np.array(pairs)
pairs = pairs[order]
ordered_pairs = np.r_[pairs,pairs,pairs,pairs]


print("Cycle:")
verbal_order = [print(f"blank for {b} s, then {s} for 3 s; ") for s,b in pairs]
print("Cycle in volumes:")
verbal_order = [print(f"blank for {b/3} volumes, then {s} for 1 volume; ") for s,b in pairs]

cycle_time = np.sum(pairs[:,1]) + len(stim)*3
print( f'Cycle total time:\n {cycle_time}, s\n {cycle_time/60}, min\n {cycle_time/3}, volumes')

plot_cycle(pairs,'One Cycle')
plot_cycle(ordered_pairs,'4 repeating cycles')
```
